```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns

codi_men_01 = pd.read_csv('codimap_men.csv')
codi_men_02 = pd.read_csv('codimap_men01.csv')
codi_men_03 = pd.read_csv('codimap_men02.csv')


dataframes = [codi_men_01, codi_men_02, codi_men_03]

codi_men_df = pd.concat(dataframes, ignore_index=True)
codi_men_df.head()
```

```python
import re

def extract_number(url):
    match = re.search(r'\d+', url)
    if match:
        return match.group()

# 'codi_img' 열에서 숫자 추출하여 'codi_img_name' 열 생성
codi_men_df['codi_img_name'] = codi_men_df['codi_img'].apply(extract_number)

codi_men_df.info()
```

```python
codi_men_df.head()
```

```python
from PIL import Image
import requests
from io import BytesIO

def save_image(url, filename):
    try:
        response = requests.get(url)
        if response.status_code == 200:
            img = Image.open(BytesIO(response.content))
            img.save(filename)
            print(f"Image saved: {filename}")
        else:
            print(f"Failed to retrieve image from URL: {url}, Status code: {response.status_code}, Content: {response.content}")
    except Exception as e:
        print(f"Error processing image from URL: {url}, Error: {str(e)}")

# 'codi_img' 열의 이미지를 저장
for index, row in codi_men_df.iterrows():
    img_url = row['codi_img']
    image_number = row['codi_img_name']
    
    if image_number:
        filename = f'codi_img_men/codi_{image_number}.jpg'  # 파일명을 적절히 지정할 수 있도록 설정
        save_image(img_url, filename)
    else:
        print(f"이미지 URL에서 숫자를 추출할 수 없습니다: {img_url}")
```
