```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns

codi_women_01 = pd.read_csv('codimap_women.csv')
codi_women_02 = pd.read_csv('codimap_women01.csv')
codi_women_03 = pd.read_csv('codimap_women02.csv')
codi_women_04 = pd.read_csv('codimap_women03.csv')

dataframes = [codi_women_01, codi_women_02, codi_women_03, codi_women_04]

codi_women_df = pd.concat(dataframes, ignore_index=True)
codi_women_df.head()
```

```python
import re

def extract_number(url):
    match = re.search(r'\d+', url)
    if match:
        return match.group()

# 'codi_img' 열에서 숫자 추출하여 'codi_img_name' 열 생성
codi_women_df['codi_img_name'] = codi_women_df['codi_img'].apply(extract_number)

codi_women_df.info()
```
